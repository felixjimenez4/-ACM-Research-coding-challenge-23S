```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd 
df = pd.read_csv("6 class csv.csv")
df
```

```python
df.head()
```

```python
stars = pd.DataFrame()
for key in df[['Temperature (K)','Luminosity(L/Lo)','Radius(R/Ro)','Absolute magnitude(Mv)']]:
    stars[key] = df.groupby('Star type')[key].mean().tolist()

stars
```

```python
stars['Temperature (K)'].plot(kind ='bar',colormap ='coolwarm')
```

```python
stars['Luminosity(L/Lo)'].plot(kind ='bar',logy = True)
```

```python
stars['Radius(R/Ro)'].plot(kind ='bar', logy = True)
```

```python
stars['Absolute magnitude(Mv)'].plot(kind = 'bar')
```

```python
df['Star color'].value_counts().plot(kind = 'pie')
```

# The most average star with

```python
avg_star = pd.DataFrame();
for key in df[['Temperature (K)','Luminosity(L/Lo)','Radius(R/Ro)','Absolute magnitude(Mv)']]:
    avg_star[key] = df[key].mean()
avg_star
```

```python
avg_star_color = df.groupby(['Star color']).mean()
avg_star_color
```

# The Temperature of the average star by color

```python
avg_star_color["Temperature (K)"].sort_values().plot( kind = ("bar"))
```

```python
avg_star_color["Luminosity(L/Lo)"].sort_values().plot( kind = ("bar"), logy = True)
```

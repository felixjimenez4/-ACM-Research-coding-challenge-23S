```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd 
df = pd.read_csv("/kaggle/input/star-dataset/6 class csv.csv")
df
```

```python
df.head()
```

```python
stars = pd.DataFrame()
for key in df[['Temperature (K)','Luminosity(L/Lo)','Radius(R/Ro)','Absolute magnitude(Mv)']]:
    stars[key] = df.groupby('Star type')[key].mean().tolist()

stars
```

```python
stars['Temperature (K)'].plot(kind ='bar',colormap ='coolwarm')
```

```python
stars['Luminosity(L/Lo)'].plot(kind ='bar',logy = True)
```

```python
stars['Radius(R/Ro)'].plot(kind ='bar', logy = True)
```

```python
stars['Absolute magnitude(Mv)'].plot(kind = 'bar')
```
